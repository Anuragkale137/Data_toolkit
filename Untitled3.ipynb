```python
#3 Explain the following terms
#M The difference in npYarray, npYasarray and npYasanyarrayX
#M The difference between Deep copy and shallow copyX
#answer-->
#np.array-This function creates a NumPy array from a list or tuple. It requires that the input is a uniform type (i.e., all elements should ideally be of the same type).
#np.asarray-Similar to np.array, but it will not create a copy if the input is already an array (or an array-like object). It simply returns the input as an array if possible.
#np.asanarray-This function is like np.asarray, but it allows input to be converted to any array type, including subclasses of ndarray (like masked arrays or matrices).
#deep copy-A deep copy creates a new object and recursively adds copies of nested objects found in the original, meaning that the new object is fully independent of the original.
#shallow copy-A shallow copy creates a new object but inserts references into it to the objects found in the original. If the original object contains references to mutable objects, those mutable objects are shared between the original and the shallow copy.

```

```python
#Kçg Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
plt.figure()
x=np.random.randint(110,size=10)
y=np.random.randint(110,size=10)
df=pd.DataFrame({'X axis': x, 'Y axis': y})


df['Quadrant'] = 0
df.loc[(df['X axis'] > 0) & (df['Y axis'] > 0), 'Quadrant'] = 1  # Quadrant 1
df.loc[(df['X axis'] < 0) & (df['Y axis'] > 0), 'Quadrant'] = 2  # Quadrant 2
df.loc[(df['X axis'] < 0) & (df['Y axis'] < 0), 'Quadrant'] = 3  # Quadrant 3
df.loc[(df['X axis'] > 0) & (df['Y axis'] < 0), 'Quadrant'] = 4  # Quadrant 4
sb.scatterplot(x='X axis',y='Y axis',hue='Quadrant',data=df,palette='viridis')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Quadrant-wise Scatter Plot")
plt.show()
```
